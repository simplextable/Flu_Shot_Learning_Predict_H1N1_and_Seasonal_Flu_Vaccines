```python
y1 = df2[["h1n1_vaccine"]]
y2 = df2[["seasonal_vaccine"]]
```

```python
x_train, x_val ,y_train, y_val = train_test_split(df1 ,y1 , test_size = 0.33 , random_state = 0)
```

```python
#opinion_h1n1_risk: 0.32
#opinion_h1n1_vacc_effective: 0.27
#doctor_recc_h1n1: 0.39
x_train = x_train[["opinion_h1n1_risk","opinion_h1n1_vacc_effective","doctor_recc_h1n1"]]
```

```python
x_train.fillna(x_train.mean(), inplace=True)
```

```python
x_val = x_val[["opinion_h1n1_risk","opinion_h1n1_vacc_effective","doctor_recc_h1n1"]]
```

```python
x_val.fillna(x_val.mean(), inplace=True)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators = 10, criterion = "entropy")
rfc.fit(x_train, y_train)
y_pred = rfc.predict(x_val)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_val,y_pred)
print("Random Forest")
print(cm)  
```
