```python
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.ensemble import GradientBoostingClassifier

pd.set_option("display.max_rows",9000)
pd.set_option("display.max_columns",50)

from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
df1 = pd.read_csv("training_set_features.csv")
df2 = pd.read_csv("training_set_labels.csv")
```

```python
#correlations for h1n1 :
#doctor_recc_h1n1: 0.39
#opinion_h1n1_risk: 0.32
#opinion_h1n1_vacc_effective: 0.27
#opinion_seas_risk: 0.26
#health_insurance : 0.22
#doctor_recc_seasonal: 0.21

#correlations for seasonal_vaccine:
#opinion_seas_risk: 0.39
#doctor_recc_seasonal: 0.37
#opinion_seas_vacc_effective: 0.36
#opinion_h1n1_risk: 0.22
#opinion_h1n1_vacc_effective: 0.21
#doctor_recc_h1n1: 0.2 
#health_insurance: 0.2

```

```python
y = df2[["h1n1_vaccine", "seasonal_vaccine"]]

x_train, x_val ,y_train, y_val = train_test_split(df1 ,y , test_size = 0.33 , random_state = 0)

x_train.fillna(x_train.mean(), inplace=True)
x_val.fillna(x_val.mean(), inplace=True)
```

```python
#The highest correlations for h1n1_vaccine 

#doctor_recc_h1n1: 0.39
#opinion_h1n1_risk: 0.32
#opinion_h1n1_vacc_effective: 0.27

x_train1 = x_train[["opinion_h1n1_risk","opinion_h1n1_vacc_effective","doctor_recc_h1n1"]]
x_val1 = x_val[["opinion_h1n1_risk","opinion_h1n1_vacc_effective","doctor_recc_h1n1"]]

#The highest correlations for seasonal_vaccine 

#opinion_seas_risk: 0.39
#doctor_recc_seasonal: 0.37
#opinion_seas_vacc_effective: 0.36

x_train2 = x_train[["doctor_recc_seasonal","opinion_seas_risk","opinion_seas_vacc_effective"]]
x_val2 = x_val[["doctor_recc_seasonal","opinion_seas_risk", "opinion_seas_vacc_effective"]]
```

```python
models=[]
h1n1_accuracy=[]
seasonal_accuracy=[]
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators = 10, criterion = "entropy")
rfc.fit(x_train1, y_train[["h1n1_vaccine"]])
y_pred1 = rfc.predict(x_val1)

from sklearn.metrics import confusion_matrix
cm1 = confusion_matrix(y_val[["h1n1_vaccine"]],y_pred1)
print("Random Forest")
print(cm1)  

h1n1_accuracy_score = accuracy_score(y_val["h1n1_vaccine"], y_pred1)
print(h1n1_accuracy_score)
models.append("RandomForest")
h1n1_accuracy.append(h1n1_accuracy_score)


#XGBoost - h1n1_vaccine
xgb1 = XGBClassifier()
xgb1.fit(x_train1, y_train["h1n1_vaccine"])
y_pred_xgb1 = xgb1.predict(x_val1) 

cm_xgb1 = confusion_matrix(y_val["h1n1_vaccine"],y_pred_xgb1)
print("XGBoost")
print(cm_xgb1) 

h1n1_accuracy_score = accuracy_score(y_val["h1n1_vaccine"], y_pred_xgb1)
print(h1n1_accuracy_score)
models.append("XGBoost")
h1n1_accuracy.append(h1n1_accuracy_score)


#Gradient Descent - h1n1_vaccine
gbc1 = GradientBoostingClassifier()
gbc1.fit(x_train1, y_train["h1n1_vaccine"])
y_pred_gbc1 = gbc1.predict(x_val1)

cm_gbc1 = confusion_matrix(y_val["h1n1_vaccine"],y_pred_gbc1)
print("GradientBoositngClassifier")
print(cm_gbc1) 

h1n1_accuracy_score = accuracy_score(y_val["h1n1_vaccine"], y_pred_gbc1)
print(h1n1_accuracy_score)
models.append("Gradient Descent")
h1n1_accuracy.append(h1n1_accuracy_score)


# Logistic Regression - h1n1_vaccine

from sklearn.linear_model import LogisticRegression
log1 = LogisticRegression(random_state=0)
log1.fit(x_train1, y_train["h1n1_vaccine"])
y_pred_log1 = log1.predict(x_val1)

cm_log1 = confusion_matrix(y_val["h1n1_vaccine"],y_pred_log1)
print("Logistic Regression")
print(cm_log1) 

h1n1_accuracy_score = accuracy_score(y_val["h1n1_vaccine"], y_pred_log1)
print(h1n1_accuracy_score)
models.append("Logistic")
h1n1_accuracy.append(h1n1_accuracy_score)


# KNN - h1n1_vaccine
from sklearn.neighbors import KNeighborsClassifier
knn1 = KNeighborsClassifier(n_neighbors=5)
knn1.fit(x_train1, y_train["h1n1_vaccine"])
y_pred_knn1 = knn1.predict(x_val1)

cm_knn1 = confusion_matrix(y_val["h1n1_vaccine"],y_pred_knn1)
print("KNN")
print(cm_knn1) 

h1n1_accuracy_score = accuracy_score(y_val["h1n1_vaccine"], y_pred_knn1)
print(h1n1_accuracy_score)
models.append("KNN")
h1n1_accuracy.append(h1n1_accuracy_score)
```

```python
#Random Forest - seasonal_vaccine
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators = 10, criterion = "entropy")
rfc.fit(x_train2, y_train["seasonal_vaccine"])
y_pred2 = rfc.predict(x_val2)

from sklearn.metrics import confusion_matrix
cm2 = confusion_matrix(y_val["seasonal_vaccine"],y_pred2)
print("Random Forest")
print(cm2) 

seasonal_accuracy_score = accuracy_score(y_val["seasonal_vaccine"], y_pred2)
print(seasonal_accuracy_score)
seasonal_accuracy.append(seasonal_accuracy_score)
```

```python
#XGBoost - seasonal_vaccine
xgb2 = XGBClassifier()
xgb2.fit(x_train2, y_train["seasonal_vaccine"])
y_pred_xgb2 = xgb2.predict(x_val2) 

cm_xgb2 = confusion_matrix(y_val["seasonal_vaccine"],y_pred_xgb2)
print("XGBoost")
print(cm_xgb2)

seasonal_accuracy_score = accuracy_score(y_val["seasonal_vaccine"], y_pred_xgb2)
print(seasonal_accuracy_score)
seasonal_accuracy.append(seasonal_accuracy_score)
```

```python
#Gradient Boosting - seasonal_vaccine
gbc2 = GradientBoostingClassifier()
gbc2.fit(x_train2, y_train["seasonal_vaccine"])
y_pred_gbc2 = gbc2.predict(x_val2)

cm_gbc2 = confusion_matrix(y_val["seasonal_vaccine"],y_pred_gbc2)
print("GradientBoositngClassifier")
print(cm_gbc2) 

seasonal_accuracy_score = accuracy_score(y_val["seasonal_vaccine"], y_pred_gbc2)
print(seasonal_accuracy_score)
seasonal_accuracy.append(seasonal_accuracy_score)
```

```python
# Logistic Regression - seasonal_vaccine
log2 = LogisticRegression(random_state=0)
log2.fit(x_train2, y_train["seasonal_vaccine"])
y_pred_log2 = log2.predict(x_val2)

cm_log2 = confusion_matrix(y_val["seasonal_vaccine"],y_pred_log2)
print("Logistic Regression")
print(cm_log2) 

seasonal_accuracy_score = accuracy_score(y_val["seasonal_vaccine"], y_pred_log2)
print(seasonal_accuracy_score)
seasonal_accuracy.append(seasonal_accuracy_score)
```

```python
# KNN - seasonal_vaccine
from sklearn.neighbors import KNeighborsClassifier
knn2 = KNeighborsClassifier(n_neighbors=5)
knn2.fit(x_train2, y_train["seasonal_vaccine"])
y_pred_knn2 = knn2.predict(x_val2)

cm_knn2 = confusion_matrix(y_val["seasonal_vaccine"],y_pred_knn2)
print("KNN")
print(cm_knn2) 

seasonal_accuracy_score = accuracy_score(y_val["seasonal_vaccine"], y_pred_knn2)
print(seasonal_accuracy_score)
seasonal_accuracy.append(seasonal_accuracy_score)
```

```python
models_accuracy_scores = pd.DataFrame(
    {'Models': models,
     'h1n1_accuracy': h1n1_accuracy,
     'seasonal_accuracy': seasonal_accuracy
    })
```
